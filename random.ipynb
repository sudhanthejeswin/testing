```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion='squared_error', max_features ='log2', n_estimators = 100, random_state = 0)       
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="Finalized_model_log.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open('Finalized_model_log.sav', 'rb'))
```

```python
print("Enter the following details:")
age = float(input("Age: "))
bmi = float(input("BMI: "))
children = int(input("Children: "))
smoker_yes = int(input("Smoker (1 for Yes, 0 for No): "))

```

```python
user_input = pd.DataFrame([[age, bmi, children, smoker_yes]], columns=['age', 'bmi', 'children', 'smoker_yes'])
```

```python
predicted_charges = loaded_model.predict(user_input)
```

```python
print(f"Predicted Charges: ${predicted_charges[0]:.2f}")
```
